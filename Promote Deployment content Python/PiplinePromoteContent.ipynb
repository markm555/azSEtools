```python
import sempy.fabric as fabric
import pandas as pd

# Instantiate the client
client = fabric.FabricRestClient()
```

```python
response = client.get("/v1/deploymentPipelines")
pipelines = response.json()
print(pipelines)
```

```python
# Get 
pipeline_id = "72f7b50e-ded4-49a7-a2fd-2bb0a313a200"
response = client.get(f"/v1/deploymentPipelines/{pipeline_id}/stages")
stages = response.json()
print(stages)
```

```python
source_stage_id = "8e2b0190-72ac-478a-ae02-8324314a3a6c"
target_stage_id = "446de1f5-2a01-45a8-9124-4bd5aa7c408f"
deploy_url = f"/v1/deploymentPipelines/72f7b50e-ded4-49a7-a2fd-2bb0a313a200/deploy"

payload = {
    "sourceStageId": source_stage_id,
    "targetStageId": target_stage_id,
    "note": "Deploying content from source to target stage"
}

response = client.post(deploy_url, json=payload)
print(response.json())
```
